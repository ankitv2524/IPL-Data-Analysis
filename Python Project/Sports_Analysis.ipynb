```python
bowler_overview = merge.groupby("bowler")[["total_runs","ball","win_by_wickets"]].agg({"total_runs":"sum", "ball": "count", "win_by_wickets":"count"}).sort_values(by = 'total_runs', ascending = False)
# bowler_overview['six'] = ((bowler_overview["total_runs"])/6)
bowler_overview["economy"] = ((bowler_overview["ball"])/((bowler_overview["total_runs"])/6))
bowler_graph = bowler_overview["economy"].head(20).sort_values(ascending = True)
bowler_graph.plot.bar(figsize=(15,5), color = 'orange', fontsize = 20)
plt.title("Bowlers Economy", fontsize = 50)
plt.xlabel("Bowlers",fontsize = 30)
plt.ylabel("Economy",fontsize = 30)
plt.show()
```

### Wickets taken by a bowler

```python
bowler_overview = merge.groupby("bowler")[["total_runs","ball","win_by_wickets"]].agg({"total_runs":"sum", "ball": "count", "win_by_wickets":"count"}).sort_values(by = 'total_runs', ascending = False)
bowler_graph_wickets = bowler_overview["win_by_wickets"].head(20)
bowler_graph_wickets.plot.bar(figsize=(15,5), color = 'gray', fontsize = 20)
plt.title("Bowlers Wickets", fontsize = 50)
plt.xlabel("Bowlers",fontsize = 30)
plt.ylabel("Wickets",fontsize = 30)
plt.show()
```

#### Top 5 Bowlers
